### Calculating The Total Times
Only use this once, because otherwise it will throw and error if the column is already transformed

```python
import pandas as pd

# Function to convert datetime.time to total seconds
def time_to_seconds(t):
    return t.hour * 3600 + t.minute * 60 + t.second

# Apply the conversion function to the 'total_duration' column
xc_training_data['total_seconds'] = xc_training_data['total_duration'].apply(time_to_seconds)
as_training_data['total_seconds'] = as_training_data['total_duration'].apply(time_to_seconds)

# Convert total seconds to hours
xc_total_hours = xc_training_data['total_seconds'].sum() / 3600
as_total_hours = as_training_data['total_seconds'].sum() / 3600

# Display the total hours
print("Total training XC time in hours:", xc_total_hours)
print("Total training AS time in hours:", as_total_hours)
```

### Time in Zones Aggregation

```python
import pandas as pd

# Function to convert time objects to seconds
def time_to_seconds(t):
    if isinstance(t, pd._libs.tslibs.timedeltas.Timedelta) or isinstance(t, pd.Timedelta):  # Already a timedelta
        return t.total_seconds()
    if hasattr(t, 'hour'):  # Check if it is a time object
        return t.hour * 3600 + t.minute * 60 + t.second
    return 0  # Return 0 if the value is not a valid time object

# List of zone columns
zone_columns = [f'zone_{i}_time' for i in range(1, 6)]

# Dictionary to store the total hours for each zone
zone_hours = {}

# Loop through each zone column, convert time to seconds, sum it, and convert to hours
for zone in zone_columns:
    # Create a new column for time in seconds
    xchr_training_data[f'{zone}_seconds'] = xchr_training_data[zone].apply(time_to_seconds)
    
    # Sum the seconds and convert to hours
    total_hours = xchr_training_data[f'{zone}_seconds'].sum() / 3600
    
    # Store the result in the dictionary
    zone_hours[zone] = total_hours
    
# Print the zone hours dictionary
print(zone_hours)

# Optionally return the dictionary
def get_zone_hours():
    return zone_hours

# Example usage of the function
zone_hours_dict = get_zone_hours()
print(zone_hours_dict)
```

### Calculate AS Total Time by Type

```python
import pandas as pd

# Define the function
def group_and_sum_time(data, group_column, time_column):
    """
    Groups the data by the specified column, sums the time column (in seconds), 
    and converts the result to hours.

    Parameters:
    - data: The DataFrame containing the data.
    - group_column: The column to group by.
    - time_column: The column that contains time values in seconds.

    Returns:
    - A sorted Series with the total hours for each group.
    """

    # Group by the specified column and sum the total seconds for each group
    grouped_data = data.groupby(group_column)[time_column].sum()

    # Convert the total seconds back to hours
    grouped_data_in_hours = grouped_data / 3600

    # Sort the grouped data in descending order
    sorted_data = grouped_data_in_hours.sort_values(ascending=False)

    return sorted_data
```

### Monthly Summary AS

```python
import pandas as pd

# Ensure 'start_date' is in datetime format
as_training_data['start_date'] = pd.to_datetime(as_training_data['start_date'])

# Extract the month and year from 'start_date'
as_training_data['month'] = as_training_data['start_date'].dt.to_period('M')  # 'M' stands for month

# Group by the month and sum the total training seconds for each month
monthly_summary = as_training_data.groupby('month')['total_seconds'].sum()

# Convert seconds to hours as a decimal
def convert_seconds_to_hours_decimal(seconds):
    hours_decimal = seconds / 3600  # Convert total seconds to hours
    return round(hours_decimal, 1)  # Round to 1 decimal place for cleaner output

# Apply the conversion function to the monthly summary
monthly_summary_hours = monthly_summary.apply(convert_seconds_to_hours_decimal)

# Convert to dictionary for easy access later
monthly_summary_hours_dict = {str(month): hours for month, hours in monthly_summary_hours.items()}

# Display the monthly summary in decimal hours
print(monthly_summary_hours)
```
